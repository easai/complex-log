# 対数関数を複素数まで拡張する

対数の定義域はご存じの通り正の数です。$x>0$でしか定義されません。

```python
from sympy.plotting import plot
from sympy import symbols, log

x = symbols("x")
plt = plot(log(x), (x, -10, 10))
```

ここで、複素数を考えると定義域、値域を拡張することができます。複素数まで含めれば、$log(-1)$まで定義できるのでした。

```python
from sympy import I, log, sqrt

log(-1)
```

複素数の対数の定義のほうですが、ここで注意しなければならないのはこれが多価関数だということです。複素数は$e^{2n\pi}$など掛けると同じ値です。

```python
import matplotlib.pyplot as plt
from sympy import re, im, arg, Abs, I, pi, sin, cos


def plotComplex(r, theta):
    fig = plt.figure(figsize=(4, 4))
    ax = plt.subplot(projection="polar")
    ax.plot(theta, r, marker="o", markersize=4, color="black")
    ax.plot((0, theta), (0, r), "--", color="gray")
    ax.text(theta - 0.07, r, f"z = {r}+{theta}i", fontsize=16)


p = sqrt(3) + I
r = Abs(p)
theta = arg(p)
plotComplex(r, theta)

theta += 2 * pi
plotComplex(r, theta)
```

つまり、偏角の部分は$2n\pi$を考えねばなりません。
\begin{align}
z&=r\cdot e^\omega\\
&=r\cdot e^\omega\cdot e^{2n\pi}\\
&=r\cdot e^{\theta+2n\pi}
\end{align}
そこで対数関数の定義はこうなります。
\begin{align}
\omega&=\log(r)+(\theta+2n\pi)i\\
&=\log|z| +(\arg(z)+2n\pi)i
\end{align}
そして$n=0$の場合を主値と呼びます。
\begin{align}
\omega&=\operatorname{Log}|z| +(\arg(z))i
\end{align}
ここで偏角のとれる範囲は$-\pi\lt\arg(z)\leq\pi$とします。

対数関数の実数部分を表したのが下のグラフです。

```python
from sympy.plotting import plot3d
from sympy import re

y = symbols("y")
z = x + I * y
plot3d(re(log(z)), (x, -5, 5), (y, -5, 5))
```

虚数部分は以下のようになります。

```python
from sympy import im

y = symbols("y")
z = x + I * y
plot3d(im(log(z)), (x, -5, 5), (y, -5, 5))
```

複素数まで拡張した対数の絶対値のグラフです。

```python
y = symbols("y")
z = x + I * y
plot3d(Abs(log(z)), (x, -5, 5), (y, -5, 5))
```
